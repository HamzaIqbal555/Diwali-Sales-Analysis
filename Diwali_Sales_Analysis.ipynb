```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head(10)
```

```python
df.info()
```

# Data Cleaning and Preprocessiing

```python
df.drop(['Status','unnamed1'], axis=1, inplace= True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']= df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.rename(columns= {'Marital_Status': 'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

# Exploratory Data Analysis (EDA)
## Gender

```python
ax = sns.countplot(x= 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(6,4)})
```

```python
gender_sales = df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x= 'Gender', y = 'Amount', data = gender_sales)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.*

## Age Group

```python
ax = sns.countplot(x= 'Age Group', data = df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_group_sales = df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x= 'Age Group', y = 'Amount', data = age_group_sales)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

## State

```python
statewise_order = df.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x= 'State', y = 'Orders', data = statewise_order)
```

```python
statewise_amount = df.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x= 'State', y = 'Amount', data = statewise_amount)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

## Marital_Status

```python
ax = sns.countplot(x= 'Marital_Status', data = df)

sns.set(rc={'figure.figsize': (7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
MSwise_amount = df.groupby(['Marital_Status','Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(6,4)})
sns.barplot(x= 'Marital_Status', y = 'Amount', data = MSwise_amount, hue = 'Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

## Occupation

```python
ax = sns.countplot(x= 'Occupation', data = df)

sns.set(rc={'figure.figsize': (17,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Occupationwise_amount = df.groupby(['Occupation','Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(16,4)})
sns.barplot(x= 'Occupation', y = 'Amount', data = Occupationwise_amount, hue = 'Gender')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

## Product_Category

```python
ax = sns.countplot(x= 'Product_Category', data = df)

# sns.set(rc={'figure.figsize': (20,5)})
plt.xticks(rotation=45, ha='right')

# Optionally, adjust the layout to prevent clipping of tick-labels
# plt.tight_layout()
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Product_Category_amount = df.groupby(['Product_Category'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

# sns.set(rc={'figure.figsize':(16,4)})
sns.barplot(x= 'Product_Category', y = 'Amount', data = Product_Category_amount)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

## Product_ID

```python
Product_Id_amount = df.groupby(['Product_ID'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

# sns.set(rc={'figure.figsize':(16,4)})
sns.barplot(x= 'Product_ID', y = 'Amount', data = Product_Id_amount)
```

## Conclusion

*Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
